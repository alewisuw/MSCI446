```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python
df1 = pd.read_csv("data/test.csv") #set from kaggle came pre-split, although we want to combine them and do our own test train split later
df2 = pd.read_csv("data/train.csv")
```

```python
df = pd.concat([df1,df2])
df.head()
```

```python
label_encoder = LabelEncoder()
ordinal_encoder = OrdinalEncoder()

df['transmission_encoded'] = label_encoder.fit_transform(df['transmission'])
```

```python
mean_sale_price = df.groupby('make')['sale_price'].mean()
sorted_make = mean_sale_price.sort_values().index
ordinal_labels = {make: i for i, make in enumerate(sorted_make)}
df['make_encoded'] = df['make'].map(ordinal_labels)

ordinal_labels
```

```python
mean_sale_price = df.groupby('car_rating')['sale_price'].mean()
sorted_car_ratings = mean_sale_price.sort_values().index
ordinal_labels = {car_rating: i for i, car_rating in enumerate(sorted_car_ratings)}
df['car_rating_encoded'] = df['car_rating'].map(ordinal_labels)

ordinal_labels
```

```python
mean_sale_price = df.groupby('fuel_type')['sale_price'].mean()
sorted_fuel = mean_sale_price.sort_values().index
ordinal_labels = {fuel_type: i for i, fuel_type in enumerate(sorted_fuel)}
df['fuel_type_encoded'] = df['fuel_type'].map(ordinal_labels)

ordinal_labels
```

```python
mean_sale_price = df.groupby('body_type')['sale_price'].mean()
sorted_body_type = mean_sale_price.sort_values().index
ordinal_labels = {body_type: i for i, body_type in enumerate(sorted_body_type)}
df['body_type_encoded'] = df['body_type'].map(ordinal_labels)

ordinal_labels
```

```python
#the columns dropped below have either been encoded above, or are specific to the website's dataset and thus not useful for our problem
df.drop(columns=['make'], inplace=True)
df.drop(columns=['model'], inplace=True)
df.drop(columns=['rto'], inplace=True)
df.drop(columns=['fuel_type'], inplace=True)
df.drop(columns=['body_type'], inplace=True)
df.drop(columns=['car_rating'], inplace=True)
df.drop(columns=['city'], inplace=True)
df.drop(columns=['registered_city'], inplace=True)
df.drop(columns=['registered_state'], inplace=True)
df.drop(columns=['car_name'], inplace=True)
df.drop(columns=['variant'], inplace=True)
df.drop(columns=['source'], inplace=True)
df.drop(columns=['car_availability'], inplace=True)
df.drop(columns=['ad_created_on'], inplace=True)
df.drop(columns=['transmission'], inplace=True)
df.drop(columns=['is_hot'], inplace=True)
df.drop(columns=['emi_starts_from'], inplace=True)
df.drop(columns=['broker_quote'], inplace=True)
df.drop(columns=['times_viewed'], inplace=True)
df.drop(columns=["id"], inplace=True)
df.drop(columns=["assured_buy"], inplace=True)
df.drop(columns=["booking_down_pymnt"], inplace=True)
df.drop(columns=["reserved"], inplace=True)
```

```python
df = df.dropna(axis=0)  # Remove rows with NaNs
len(df)
```
